```python
agenda = {}

def adicionar_contato():
  nome = input("Digite o nome do novo contato: ")
  telefone = input("Digite o telefone: ")
  agenda[nome] = telefone
  print("Contato adicionado com sucesso!")

def buscar_contato():
  nome = input("Digite o nome do contato que deseja buscar: ")
  if nome in agenda:
    print(f"{nome}: {agenda[nome]}")
  else:
    print("Contato não encontrado.")

def listar_contato():
  if not agenda:
    print("Agenda Vazia.")
  else:
    print("\nContatos: ")
    for nome, telefone in agenda.items():
        print(f"{nome}\t{telefone}")

def editar_contato():
  nome = input("Digite o nome do contato que deseja editar: ")
  if nome in agenda:
    novo_telefone = input("Digite o novo telefone: ")
    agenda[nome] = novo_telefone
    print(f"Telefone de {nome} foi atualizado com sucesso!")
  else:
    print("Contato não encontrado")

def excluir_contato():
  nome = input("Digite o nome do contato que deseja excluir: ")
  if nome in agenda:
    confirmacao = input(f"Tem certeza que deseja excluir {nome}? (s/n): ")
    if confirmacao.lower() == "s":
      del agenda[nome]
      print(f"{nome} excluído com sucesso.")
    else:
      print("Exclusão cancelada.")
  else:
    print("Contato não encontrado")

while True:
  print("\n=== MENU ===")
  print("1. Listar contatos")
  print("2. Buscar contato")
  print("3. Adicionar contato")
  print("4. Editar contato")
  print("5. Excluir contato")
  print("6. Sair")

  opcao = input("Selecione a opção que deseja (De 1 a 6): ")

  if opcao == "1":
      listar_contato()
  elif opcao == "2":
      buscar_contato()
  elif opcao == "3":
      adicionar_contato()
  elif opcao == "4":
      editar_contato()
  elif opcao == "5":
    excluir_contato()
  elif opcao == "6":
    print("Saindo da agenda. Até mais tarde!")
    break
  else:
    print("Opção inválida! Tente novamente (Apenas de 1 a 6)")
```
