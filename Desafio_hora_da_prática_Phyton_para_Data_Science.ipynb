```python
from random import choice
frutas = ["maçã", "banana", "uva", "pêra",
          "manga", "coco", "melancia", "mamão",
          "laranja", "abacaxi", "kiwi", "ameixa"]
contador = 0
salada = []
while contador != 3:
  salada.append(choice(frutas))
  contador += 1
print(f'Sua salada terá as frutas: {', '.join(salada)}')
```
