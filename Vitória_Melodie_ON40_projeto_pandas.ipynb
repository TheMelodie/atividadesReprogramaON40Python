#### Lidando com dados em csv

Execute o bloco de código a seguir para transformar os dados csv em Data Frame

```python
# Definindo o caminho para buscar a base de dados
baseDadosDeputadosCsv = '/content/deputados_legislatura_57_base_tratada.csv'

# Transformando a base de dados de csv para o formato Data Frame
dfDeputadosCsv = pd.read_csv(baseDadosDeputadosCsv)

#selecionar apenas algumas colunas
colunasCsv = ['Nome', 'Estado', 'Sigla Gênero', 'Gênero', 'Data de nascimento', 'Idade', 'Primeira legislatura']
dfDeputadosCsv = dfDeputadosCsv[colunasCsv]

# Exibir as colunas do dataframe
print(dfDeputadosCsv.columns)

# Exibir as quantidade de linhas
dfDeputadosCsv.value_counts()
```

```python
dfDeputadosCsv
```

#### Lidando com dados em json

Execute o bloco de código a seguir para transformar os dados json em Data Frame


```python
import json

# Definindo o caminho para buscar a base de dados
baseDadosDeputadosJson = '/content/deputados_legislatura_57.json'

# Transformando a base de dados de json para o formato Data Frame
dfDeputadosJson = pd.read_json(baseDadosDeputadosJson)

#selecionar apenas algumas colunas
colunasJson = ['id', 'nome', 'siglaPartido', 'email']
dfDeputadosJson = dfDeputadosJson[colunasJson]

# Exibir as colunas do dataframe
print(dfDeputadosJson.columns)

# Exibir as quantidade de linhas
print(dfDeputadosJson.value_counts)

# Imprime os as colunas do dataframe
print(dfDeputadosJson.columns)
print(dfDeputadosJson.value_counts)
```

#### Juntar bases de dados em uma só

Vamos utilizar o conceito de [joins (ou junção)](https://https://www.alura.com.br/artigos/join-e-seus-tipos?srsltid=AfmBOorqiWzXLMyZKkskHWph08lQWzlGAOAJJ-DQR-cC4DF46cQ_VnA1 ) para mesclar as duas bases em uma só a partir da coluna "nome"

```python
# Juntar as duas bases num inner join usando as colunas de nome
dfDeputadosFinal = pd.merge(
    dfDeputadosJson,
    dfDeputadosCsv,
    left_on='nome',
    right_on='Nome',
    how='inner'
)

# Remover a coluna duplicada Nome
dfDeputadosFinal = dfDeputadosFinal.drop(columns=['Nome'])

print(dfDeputadosFinal.head(3))
```

3. Qual a proporção de reeleição e primeira legislatura por gênero?

```python
# Contagem cruzada
proporcaoReeleicao = pd.crosstab(dfDeputadosFinal['Gênero'], dfDeputadosFinal['Primeira legislatura'], normalize='index') * 100

pd.crosstab(dfDeputadosFinal['Gênero'], dfDeputadosFinal['Primeira legislatura'], normalize='index')

display(proporcaoReeleicao.round(1))
```

4. Quais partidos têm a maior quantidade de deputadas mulheres e qual é a proporção de mulheres em relação aos homens em cada partido?

```python
# Filtrar apenas mulheres
dfMulheres = dfDeputadosFinal[dfDeputadosFinal['Gênero'] == 'Feminino']

# Contar mulheres por partido
mulheresPartido = dfMulheres['siglaPartido'].value_counts().head(10)
```

```python
# Contar quantidade de mulheres e homens por partido
generoPartido = dfDeputadosFinal.groupby(['siglaPartido', 'Gênero']).size().unstack(fill_value=0)

# Calcular o total de deputados por partido
generoPartido['Total'] = generoPartido.sum(axis=1)

# Calcular a porcentagem de mulheres em relação ao total
generoPartido['% Mulheres'] = (generoPartido['Feminino'] / generoPartido['Total']) * 100

# Ordenar pelos partidos com mais mulheres
resultado = generoPartido.sort_values(by='Feminino', ascending=False)

# Mostrar as 10 primeiras linhas
resultado.head(10)
```

```python
# Formatar a coluna de porcentagem com duas casas decimais e o símbolo %
resultado['% Mulheres'] = resultado['% Mulheres'].apply(lambda x: f'{x:.2f}%')

# Exibir novamente
resultado.head(10)
```

5. Faça um inner join e conte quantos nomes ficaram de fora do CSV e do JSON. → Dica: compare os tamanhos antes e depois do merge.

```python
import pandas as pd

# Conto o número de linhas nos DataFrames originais
linhasCsvOriginal = len(dfDeputadosCsv)
linhasJsonOriginal = len(dfDeputadosJson)

# Removo duplicatas do dfDeputadosFinal com base no 'id' antes de contar para precisar a conatgem #
dfDeputadosUnicos = dfDeputadosFinal.drop_duplicates(subset=['id'])

# Conto o número de linhas no DataFrame após o inner join e remoção de duplicatas #
totalParlamentaresUnicos = len(dfDeputadosUnicos)

# Calculo quantos nomes ficaram de fora de cada base original #
nomesForaCsv = linhasCsvOriginal - totalParlamentaresUnicos
nomesForaJson = linhasJsonOriginal - totalParlamentaresUnicos

# Crio uma Series para exibir os resultados de forma organizada #
resultadosAnalise = pd.Series({
    'Parlamentares no CSV original': linhasCsvOriginal,
    'Parlamentares no JSON original': linhasJsonOriginal,
    'Parlamentares únicos após o inner join': totalParlamentaresUnicos,
    'Nomes do CSV que não estão no inner join': nomesForaCsv,
    'Nomes do JSON que não estão no inner join': nomesForaJson
})

display(resultadosAnalise)
```
