```python
qtdColoniaA = 4
qtdColoniaB = 10
dias = 0
def calculardias (popA, popB, tempo):
  while popA < popB:
    popA = popA + (popA*0.03)
    popB = popB + (popB*(0.015))
    tempo += 1
  return print(f'População A: {popA},\nPopulação B: {popB}.\nLevou {tempo} dias.')
calculardias(qtdColoniaA, qtdColoniaB, dias)
```

```python
for i in range (1, 16):
  print(f'Usuário {i}')
  nota = input('\nInsira a nota de 0 a 5: ')
  while type(nota) == str:
    print('Inserção Inválida!')
    nota = input(f'\nUsuário {i}\n\nInsira um valor válido! (De 0 a 5): ')
  notaConverted = float(nota)
  while notaConverted < 0 or notaConverted > 5:
    print('\nNota inválida!')
    notaConverted = float(input(f'\nUsuário {i}\n\nInsira um valor válido! (De 0 a 5): '))
  print(f'\nUsuário {i}, Nota dada: {notaConverted}\n')
```

```python
comprasmaisde3000 = 0
totalcompras = 0
lista = [2172.54, 3701.35, 3518.09, 3456.61, 3249.38, 2840.82, 3891.45, 3075.26, 2317.64, 3219.08]

for compra in lista:
  if compra > 3000.00:
    comprasmaisde3000 += 1
    porcentagem = int((comprasmaisde3000/len(lista))*100)
print(f'O total foi de {comprasmaisde3000} compras acima de 3000, um percentual de {porcentagem}% em relação a todas as compras.')
```
