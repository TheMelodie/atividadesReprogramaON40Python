Análise de eficiência no tempo de entrega

```python
tempo_prometido_bebidas = 30
tempo_prometido_sobremesas = 45
tempo_real_bebidas = 35
tempo_real_sobremesas = 40
entrega_no_prazo_bebidas = False
entrega_no_prazo_sobremesas = False

if tempo_real_bebidas <= tempo_prometido_bebidas:
    entrega_no_prazo_bebidas = True
else:
    entrega_no_prazo_bebidas = False

if tempo_real_sobremesas <= tempo_prometido_sobremesas:
    entrega_no_prazo_sobremesas = True
else:
    entrega_no_prazo_sobremesas = False

if entrega_no_prazo_bebidas == True:
    print("Está cumprindo o tempo prometido de entrega de bebidas.")
else:
    print("Não está cumprindo o tempo prometido de entrega de bebidas.")

if entrega_no_prazo_sobremesas == True:
    print("Está cumprindo o tempo prometido de entrega de sobremesas.")
else:
    print("Não está cumprindo o tempo prometido de entrega de sobremesas.")
```
