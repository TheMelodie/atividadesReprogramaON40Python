#### **ExercÃ­cio 2**

Receba duas variÃ¡veis, converta para nÃºmero e retorne o maior valor

```python
# @title
# ExercÃ­cio 2
valor1 = int(input('Insira o primeiro valor: '))
valor2 = int(input('Insira o segundo valor: '))
def retornarMaiorValor(input1: int=0, input2: int=0):
  if input1 > input2:
    return input1
  if input1 < input2:
    return input2
  else:
    return 'Os valores sÃ£o iguais'

# Exibindo o resultado do exercÃ­cio 2
print(retornarMaiorValor(valor1, valor2))
```

#### **ExercÃ­cio 3**
Receba dois valores booleanos, verifique se ambos sÃ£o verdaadeiros (True)

```python
# ExercÃ­cio 3
valorBool1 = bool(input('Insira o primeiro booleano: '))
valorBool2 = bool(input('Insira o segundo booleano: '))
def valoresSaoVerdadeiros(valor1: bool=False,valor2: bool=False):
  if valor1 == False or valor2 == False:
    return False
  else:
    return True

# Exibindo o resultado do exercÃ­cio 3
print(valoresSaoVerdadeiros(valorBool1, valorBool2))
```

```python
# Testes exercÃ­cio 3
assert valoresSaoVerdadeiros(False, False) == False
assert valoresSaoVerdadeiros(False, True) == False
assert valoresSaoVerdadeiros(True, True) == True
print('Teste passou!')
```

#### **ExercÃ­cio 5**

Com base no dicionÃ¡rio a seguir, encontre e retorne uma lista de todas as chaves que tem o valor `True`

```python
# ExercÃ­cio 5
valor = True
dadosLoja = {
    'nome': 'Loja de Livros FantÃ¡sticos',
    'faturamento': 150000.00,
    'estoque_baixo': True,
    'cidades_atendidas': ['SÃ£o Paulo', 'Curitiba', 'CuiabÃ¡', 'Manaus', 'Recife'],
    'meta_atingida': False,
    'ano_fundacao': 2010,
    'loja_online': True
}
def buscarValor(inputDict, inputValor):
  keysRetornadas = []
  for key, value in inputDict.items():
    if value == inputValor:
      keysRetornadas.append(key)
  return keysRetornadas

# Exibindo o resultado do exercÃ­cio 5
print(buscarValor(dadosLoja, valor))
```

```python
dicionarioTeste = { 'valor1': True, 'valor2': False}
assert buscarValor(dicionarioTeste, True) == ['valor1']
print('Teste passou!')
```
